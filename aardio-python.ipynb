## 简短代码使用 python 和 aardio 开发计算器并打包输出

```python
class Calculator:
    def __init__(self, mainForm):
        self.mainForm = mainForm
        self.current = ""

    def add(self, number_or_operation):
        self.current += number_or_operation
        self.mainForm.edit.text = self.current

    def clear(self, is_all=False):
        if is_all:
            self.current = ""
            self.mainForm.edit.text = self.current
            return

        self.current = self.current[:-1]
        self.mainForm.edit.text = self.current
        return

    def equal(self):
        result = eval(self.current)
        self.mainForm.edit.text = result
```

## 两种方式开发 python 百度翻译爬虫界面+打包 

```python
import requests
import random
import json
import joblib
from hashlib import md5


class BaiDuFanyi:
    def __init__(self, appKey, appSecret):
        self.url = "https://fanyi.api.baidu.com/api/trans/vip/translate"
        self.appid = appKey
        self.secretKey = appSecret
        self.fromLang = "auto"
        self.toLang = "zh"
        self.salt = random.randint(32768, 65536)
        self.header = {"Content-Type": "application/x-www-form-urlencoded"}

    def BdTrans(self, text):
        sign = self.appid + text + str(self.salt) + self.secretKey
        md = md5()
        md.update(sign.encode(encoding="utf-8"))
        sign = md.hexdigest()
        data = {
            "appid": self.appid,
            "q": text,
            "from": self.fromLang,
            "to": self.toLang,
            "salt": self.salt,
            "sign": sign,
        }
        response = requests.post(self.url, params=data, headers=self.header)
        text = response.json()
        print(text)
        results = text["trans_result"][0]["dst"]
        return results


def trans(mainForm, need_to_trans):
    content = joblib.load("password")
    appKey = content[0]
    appSecret = content[1]
    BaiduTranslate_test = BaiDuFanyi(appKey, appSecret)
    Results = BaiduTranslate_test.BdTrans(need_to_trans)
    print(Results)
    mainForm.edit2.text = Results
```
